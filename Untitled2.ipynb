```python
from elasticsearch import Elasticsearch  

# elasticsearch集群服务器的地址
ES = [
    '192.168.1.2:9200'
]

# 创建elasticsearch客户端
es = Elasticsearch(
    ES,
    # 启动前嗅探es集群服务器
    sniff_on_start=True,
    # es集群服务器结点连接异常时是否刷新es节点信息
    sniff_on_connection_fail=True,
    # 每60秒刷新节点信息
    sniffer_timeout=60
)
```

```python
import json
```

```python
filepath = r'C:/Users/SWB/PycharmProjects/scrapy-demo/demo1/logs.txt'
datas= []
with open(filepath,'r',encoding='utf-8') as f:
    for line in f.readlines():
        datas.append(json.loads(line))
```

```python
datas[1]
```

```python
print(len(datas))
```

```python
for data in datas:
    if data['date']!=None:
        data['date'] = data['date'][0:5]
```

```python

mapping = {
    'properties': {
        'title': {
            'type': 'text',
            'analyzer': 'ik_max_word',
            'search_analyzer': 'ik_max_word'
        },
        'date':{
            'type':'keyword'
        },
        'content':{
            'type':'text',
            'analyzer': 'ik_max_word',
            'search_analyzer': 'ik_max_word'
        }
    }
}
```

```python
es.indices.delete(index='loldata', ignore=[400, 404])
es.indices.create(index='loldata', ignore=400)
result = es.indices.put_mapping(index='loldata', doc_type='article', body=mapping)
print(result)
```

```python
es.index(index='loldata', doc_type='article', body=datas[1])
```

```python
for data in datas:
    try:
        es.index(index='loldata', doc_type='article', body=data)
    except:
        print(data)
```

```python
result = es.search(index='loldata', doc_type='article')
print(result)
```

```python
ES = [
    '192.168.1.2:9200'
]

# 创建elasticsearch客户端
es = Elasticsearch(
    ES,
    # 启动前嗅探es集群服务器
    sniff_on_start=True,
    # es集群服务器结点连接异常时是否刷新es节点信息
    sniff_on_connection_fail=True,
    # 每60秒刷新节点信息
    sniffer_timeout=60
)
```

```python
dsl = {
    'query': {
        'match': {
            'title': '中国'
        }
    }
}

result = es.search(index='loldata', doc_type='article', body=dsl)
print(json.dumps(result, indent=2, ensure_ascii=False))
```
